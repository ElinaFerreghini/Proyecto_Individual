```python
import pandas as pd

def transform_dataset(dataset_path, platform_name):
    # Leer el dataset desde un archivo csv
    df = pd.read_csv(dataset_path)
    
    # Crear la columna "id" usando la primera letra del nombre de la plataforma
    df["id"] = platform_name[0] + df["show_id"].astype(str)
    
    # Reemplazar los valores nulos de "rating" por "G"
    df["rating"].fillna("G", inplace=True)
    
    # Convertir las fechas a formato AAAA-mm-dd
    df["date_added"] = pd.to_datetime(df["date_added"]).dt.strftime("%Y-%m-%d")
    
    # Convertir la columna "duration" en dos columnas "duration_int" y "duration_type"
    df[["duration_int", "duration_type"]] = df["duration"].str.extract(r"(\d+) (\D+)")
    
    # Convertir los valores de los campos de texto a minúsculas
    df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
    
    return df
```

```python
amazon_prime_df = transform_dataset("MLOpsReviews/amazon_prime_titles.csv", "amazon_prime")
disney_plus_df = transform_dataset("MLOpsReviews/disney_plus_titles.csv", "disney_plus")
hulu_df = transform_dataset("MLOpsReviews/hulu_titles.csv", "hulu")
netflix_df = transform_dataset("MLOpsReviews/netflix_titles.csv", "netflix")
```

```python
import glob
```

```python
# Lee los archivos CSV y los concatena en un solo DataFrame
csv_files = glob.glob('MLOpsReviews/ratings/*.csv')
df_list = []
for csv_file in csv_files:
    df_ratings = pd.read_csv(csv_file)
    df_list.append(df_ratings)
df_ratings = pd.concat(df_list)

# Agrupa el DataFrame por movieId y calcula el promedio de rating

ratings = df_ratings.groupby('movieId')['rating'].mean().to_frame()

##ACORDATE DE SACARLOS DUPLICADOS
```

```python
ratings.head()
```

```python
# Unir los DataFrames por las columnas 'id' y 'movieId'
df = pd.merge(df, ratings, left_on='id', right_on='movieId')
```
