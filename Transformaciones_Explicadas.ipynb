**Transformaciones**: Para este MVP no necesitas perfección, ¡necesitas rapidez! ⏩ Vas a hacer estas, y solo estas, transformaciones a los datos: 


1- Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

2- Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”

3- De haber fechas, deberán tener el formato AAAA-mm-dd

4- Los campos de texto deberán estar en minúsculas, sin excepciones

5- El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
import pandas as pd
```

```python
import numpy as np
```

```python
##Cargar el archivo csv
df=pd.read_csv('MLOpsReviews/amazon_prime_titles.csv')
```

```python
df.head()
```

```python
##Genero el campo id
df["id"] = "a" + df["show_id"].astype(str)  #concatenación
```

```python
df.head()
```

```python
##Reemplazo los valores nulos del campo rating por "G"

df["rating"].fillna("G", inplace=True)  #inplace=True indica que los cambios deben realizarse en el mismo dataframe
```

```python
df.head()
```

```python
# Convertir la columna date_added a datetime y luego a string con formato AAAA-mm-dd
df["date_added"] = pd.to_datetime(df["date_added"], format="%B %d, %Y").dt.strftime("%Y-%m-%d")
```

```python
df.head()
```

```python
##paso todos los campos de texto a minisculas
df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)
```

```python
df.head()
```

```python
# Convertir la columna duration a dos columnas duration_int y duration_type
df[["duration_int", "duration_type"]] = df["duration"].str.extract("(\d+) (\D+)", expand=True)
```

```python
###Este código utiliza el método str.extract() de pandas con una expresión regular para extraer la parte numérica y la parte 
# de la unidad de duración de la columna "duration". La expresión regular (\d+) (\D+) busca un grupo de uno o más dígitos seguido 
#de un espacio en blanco, seguido de un grupo de uno o más caracteres no numéricos. El parámetro expand=True indica que se deben 
# crear dos columnas a partir de los grupos extraídos.

#El resultado es un dataframe con las nuevas columnas "duration_int" y "duration_type". La columna "duration_int" contiene los valores 
# numéricos de la duración y la columna "duration_type" contiene los valores de la unidad de duración ("min" para minutos o "season" 
# para temporadas).
```
